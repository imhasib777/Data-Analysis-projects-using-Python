```python
import numpy as np
import matplotlib.pyplot as plt
```

#List of student's CGPA in the 1st year 1st semester and 2nd semester.

```python
s1=[3.29,3.59,3.52,3.89,3.29,3.09,3.56,1.82,2.86,3.59,3.80,2.50,2.52,3.49,0.00,3.80,3.51,3.30,2.64,2.75,1.70,3.22,1.40,3.26,2.93,2.08,0.00,2.97,3.31,1.64,2.56,2.45,2.22,3.76,3.20,3.91,3.56,0.00,1.42,3.59,3.10]
s2=[2.63,3.51,2.88,3.75,3.18,2.59,3.46,2.93,2.76,3.20,3.79,1.48,2.31,3.59,0.00,3.33,3.56,3.66,2.99,3.02,1.76,2.92,3.03,0.00,2.70,3.51,0.00,2.73,3.34,1.42,2.86,3.03,2.13,3.55,3.49,3.54,3.69,0.00,1.08,3.12,3.28]
s1_a=np.array(s1)
s2_a=np.array(s2)
```

#1.Average(Total) CGPA of each students

```python
total_CGPA=np.add(s1_a,s2_a)/2
id=np.arange(1,42)
result=np.column_stack((id,total_CGPA))
print(result)
```

#2.Plot diagram of the average CGPA of each students

```python
plt.figure(figsize=(7, 7))
plt.autoscale(enable=True, axis='both', tight=True)
plt.tight_layout()
plt.plot(id,total_CGPA,color='k',marker='o')
plt.xticks(np.arange(1,43,1))
plt.yticks(np.arange(0,4,0.1))
plt.xlabel("ID")
plt.ylabel("CGPA")
plt.title("CGPA vs ID")
plt.grid(True)
plt.savefig("CGPA_vs_ID.png")
```

#3.The plot diagram of comparison between the 1st and 2nd semester result of each student
Here black plot denotes 1st semester and red shows the 2nd semester

```python
plt.figure(figsize=(8, 7))
plt.autoscale(enable=True, axis='both', tight=True)
plt.plot(id,s1_a,color='k',marker='o',label="1.1")
plt.plot(id,s2_a,color='r',linestyle='--',marker='o',label='1.2')
plt.xticks(np.arange(1,44,1))
plt.yticks(np.arange(0,5,0.1))
plt.xlabel("ID")
plt.ylabel("CGPA")
plt.title("CGPA vs ID")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("CGPA_vs_ID.png")
```

#4.The bar diagram of comparison between the 1st and 2nd semester result of each student
Here black bar denotes 1st semester and red shows the 2nd semester

```python
plt.figure(figsize=(10, 8))
plt.autoscale(enable=True, axis='both', tight=True)
plt.tight_layout()
width=0.3
plt.bar(id-width,s1_a,width=width,color='k',label="1.1")
plt.bar(id,s2_a,width=width,color='r',label='1.2')
plt.xticks(np.arange(1,43,1))
plt.yticks(np.arange(0,4.5,0.1))
plt.xlabel("ID")
plt.ylabel("CGPA")
plt.title("CGPA vs ID")
plt.grid(True)
plt.legend()
plt.savefig("CGPA_vs_ID.png")
```

#10.CGPA growth of each students
Here negative values shows the negative growth and positive shows the improvement

```python
plt.figure(figsize=(8, 8))
plt.autoscale(enable=True, axis='both', tight=True)
plt.tight_layout()
width=0.7
growth=np.subtract(s2_a,s1_a)
plt.bar(id-width,growth,width=width,color='k')
plt.xticks(np.arange(0,43,1))
plt.yticks(np.arange(-3.5,2,0.1))
plt.xlabel("ID")
plt.ylabel("CGPA")
plt.title("CGPA_growth vs ID")
plt.grid(True)
plt.show()
```
