```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#List of student's CGPA in the 1st year 1st semester and 2nd semester.

```python

s1=[3.29,3.59,3.52,3.89,3.09,3.56,1.82,2.86,3.59,3.70,2.50,2.52,3.49,3.00,3.80,3.51,3.30,2.64,2.75,1.70]
s2=[2.63,3.51,3.88,3.75,3.18,2.59,3.46,2.93,2.76,3.20,3.79,3.48,2.31,3.59,3.00,3.73,3.56,3.66,2.99,3.02]
id=np.arange(1,21,dtype='int')
name=['Shopnil','Jim','Sajidur','Nokib','Nahid','Shakib','Masum','Rizvi','sowmik','Ovi','Peyal','Omi','Sajid','Omi','Joy','Naim','Sihab','Shamim','Shanto','Mahtab']
```

#Creating a Data frame

```python
df=pd.DataFrame({'Students ID':id,'Name':name,'1st_sem':s1,'2nd_sem':s2})
df
```

#Modifying Students Id

```python

df["Students ID"]="21CSE0"+df["Students ID"].astype(  str)
df
```

#Adding new column name 'Total CGPA'

```python
df["Total CGPA"]=(df["1st_sem"]+df["2nd_sem"])/2
df
```

#Adding new column named 'Growth(%)'

```python
df["Growth(%)"]=((df["2nd_sem"]-df["1st_sem"])/df["1st_sem"])*100
df["Growth(%)"]=df["Growth(%)"].round(2)
df["Growth(%)"]=df["Growth(%)"].astype(str)
df["Growth(%)"]=df["Growth(%)"]+"%"
df
```

```python
df.set_index("Students ID",inplace=True)
df
```

```python
df.reset_index(inplace=True)
df
```

#Get the details of those students who's Total CGPA is more than 3.5
Using filtering

```python
filt=(df["Total CGPA"]>3.5)
df[filt]
```

#Only show the name of those students who's Total CGPA is more than 3.5 Using filtering

```python
df.loc[filt,"Name"]
```

#2.Plot diagram of the average CGPA of each students

```python
plt.figure(figsize=(7, 7))

plt.autoscale(enable=True, axis='both', tight=True)
plt.tight_layout()

plt.plot(id,df["Total CGPA"],color='k',marker='o')

plt.xticks(np.arange(0,22,1))
plt.yticks(np.arange(2,4.4,0.1))

plt.xlabel("ID")
plt.ylabel("CGPA")
plt.title("CGPA vs ID")

plt.grid(True)
plt.savefig("CGPA_vs_ID.png")
```

#3.The plot diagram of comparison between the 1st and 2nd semester result of each student
Here black plot denotes 1st semester and red shows the 2nd semester

```python
plt.figure(figsize=(8, 7))
plt.autoscale(enable=True, axis='both', tight=True)

plt.plot(id,df["1st_sem"],color='k',marker='o',label="1.1")
plt.plot(id,df["2nd_sem"],color='r',linestyle='--',marker='o',label='1.2')

plt.xticks(np.arange(0,22,1))
plt.yticks(np.arange(1.4,4.4,0.1))

plt.xlabel("ID")
plt.ylabel("CGPA")
plt.title("CGPA vs ID")

plt.grid(True)

plt.legend()
plt.tight_layout()

plt.savefig("CGPA_vs_ID.png")
```

#4.The bar diagram of comparison between the 1st and 2nd semester result of each student
Here black bar denotes 1st semester and red shows the 2nd semester

```python
plt.figure(figsize=(10, 8))
plt.autoscale(enable=True, axis='both', tight=True)
plt.tight_layout()

width=0.3
plt.bar(id-width,df["1st_sem"],width=width,color='k',label="1.1")
plt.bar(id,df["2nd_sem"],width=width,color='r',label='1.2')


plt.xticks(np.arange(0,22,1))
plt.yticks(np.arange(0,4.5,0.1))

plt.xlabel("ID")
plt.ylabel("CGPA")
plt.title("CGPA vs ID")

plt.grid(True)
plt.legend()
plt.savefig("CGPA_vs_ID.png")
```

#10.CGPA growth of each students
Here negative values shows the negative growth and positive shows the improvement

```python
plt.figure(figsize=(8, 8))
plt.autoscale(enable=True, axis='both', tight=True)
plt.tight_layout()
width=0.7
growth=np.subtract(df["1st_sem"],df["2nd_sem"])

plt.bar(id-width,growth,width=width,color='r')

plt.xticks(np.arange(0,22,1))
plt.yticks(np.arange(-3.5,2,0.1))
plt.xlabel("ID")
plt.ylabel("CGPA")
plt.title("CGPA_growth vs ID")
plt.grid(True)
plt.show()
```
