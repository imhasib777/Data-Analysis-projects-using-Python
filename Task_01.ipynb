```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Datasets/all_data.csv")
```

```python
df.head(10)
```

```python
df.columns
```

###1. Removing NAN value

```python
nan_values=df[df.isna().any(axis=1)]
nan_values
```

```python
df=df.dropna(how="any")
df.head()
```

```python
nan_values=df[df.isna().any(axis=1)]
nan_values
```

###So all the nan values are removed

```python
df=df[df["Order Date"].str[0:2]!="Or"]
```

# **2. Creating  a new column "City"**

```python
df["City_name"]=df["Purchase Address"].str.split(",").str[1]
df.head(5)
```

#**3. Creating a new column City-State**

```python
df["City-State"]=df["Purchase Address"].str.split(",").str[1]+" "+df["Purchase Address"].str.split(",").str[2]
df.head(5)
```

#**4. Creating a month column**

```python
df["Month"]=df["Order Date"].str[0:2]
df.head(5)
```

#**5. Replacing the value of "month" column with month name**

```python
df["Month"]=df["Month"].replace(["01","02","03","04","05","06","07","08","09","10","11","12"],["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"])
```

```python
df.head(5)
```

#**6. Adding "sales" column**

```python
df["Quantity Ordered"]=pd.to_numeric(df["Quantity Ordered"])
df["Price Each"]=pd.to_numeric(df["Price Each"])
df["Sales"]=df["Quantity Ordered"]*df["Price Each"]
df.head(5)
```

#**7. What city sold the most product in the whole year?**

```python
c=df.groupby("City_name").sum()["Sales"].sort_values(ascending=False)
c
```

```python
c.head(1)
```

#**8. What sate sold the most product?**

```python
cs=df.groupby("City-State").sum()["Sales"].sort_values(ascending=False)
cs
```

```python
cs.head(1)
```

#**9. In every month which city has the most sell?**

```python

df.loc[df.groupby('Month')['Sales'].idxmax(), ['Month', 'City_name', 'Sales']]
```

#**10. Which month has the most sell?**

```python
month=df.groupby("Month").sum()["Sales"].sort_values(ascending=False)
month
```

```python
month.head(1)
```

#So the expected month is December

#**11. Adding Time(Hour) column**

```python
df["Time(Hour)"]=df["Order Date"].str[9:11]
df.head(5)
```

#**12. Which is the peak hour for sell (with respect to the whole year)?**

```python
hour=df.groupby("Time(Hour)").sum()["Sales"].sort_values(ascending=False)
hour
```

```python
hour.head(1)
```

#**19.00 to 19.59 is the peak time when  product sold the most.So this is the time when we should display advertisements.**

#**12. Which is the peak hour of selling products (In every month)?**

```python
monthly_hourly_sales = df.groupby(['Month', 'Time(Hour)'])['Sales'].sum().reset_index()
peak_hour_per_month = monthly_hourly_sales.loc[monthly_hourly_sales.groupby('Month')['Sales'].idxmax()]
peak_hour_per_month
```

#**13. What product sold the most in the whole year?**

```python
p=df.groupby("Product").sum()["Quantity Ordered"].sort_values(ascending=False)
p
```

```python
p.head(1)
```

#**14. What product sold the most in every month?**

```python
monthly_product_sales = df.groupby(['Month', 'Product'])["Quantity Ordered"].sum().reset_index()
peak_product_per_month = monthly_product_sales.loc[monthly_product_sales.groupby('Month')["Quantity Ordered"].idxmax()]
peak_product_per_month
```
